```python
import pandas as pd

# Load loan data from Excel file
loan_data = pd.read_excel("loan_data.xlsx")

# Define a function to calculate credit score
def calculate_credit_score(row):
    # Check if sum of current loans > approved limit
    if row['current_loans_sum'] > row['approved_limit']:
        return 0

    # Calculate credit score based on the mentioned components
    credit_score = (
        row['past_loans_paid_on_time'] * 0.3 +
        row['num_loans_taken'] * 0.2 +
        row['loan_activity_current_year'] * 0.2 +
        row['loan_approved_volume'] * 0.2
    )

    return credit_score

# Apply the function to each row of the DataFrame to calculate credit scores
loan_data['credit_score'] = loan_data.apply(calculate_credit_score, axis=1)

# Define a function to approve loans based on credit score and other conditions
def approve_loan(row):
    # Condition 1: If sum of all current EMIs > 50% of monthly salary, don’t approve any loans
    if row['current_emis_sum'] > 0.5 * row['monthly_salary']:
        return 'Not Approved - High EMIs'

    # Condition 2: If credit_rating > 50, approve loan
    if row['credit_score'] > 50:
        return 'Approved'
    # Condition 3: If 50 > credit_rating > 30, approve loans with interest rate > 12%
    elif 50 >= row['credit_score'] > 30:
        return 'Approved with Interest Rate > 12%'
    # Condition 4: If 30 > credit_rating > 10, approve loans with interest rate > 16%
    elif 30 >= row['credit_score'] > 10:
        return 'Approved with Interest Rate > 16%'
    # Condition 5: If 10 > credit_rating, don’t approve any loans
    elif row['credit_score'] <= 10:
        return 'Not Approved - Low Credit Score'

    return 'Not Approved - Unknown Reason'

# Apply the loan approval function to each row of the DataFrame
loan_data['loan_approval_status'] = loan_data.apply(approve_loan, axis=1)

# Display the loan data with approval status
print(loan_data[['customer_id', 'loan_approval_status']])
```
